```python
def convertir_a_minutos(tiempo):
    partes = tiempo.split(':')
    # Si hay 2 partes (HH:MM)
    if len(partes) == 2:
        horas, minutos = map(int, partes)
        return horas * 60 + minutos

    # Si hay 3 partes (HH:MM:SS)
    elif len(partes) == 3:
        horas, minutos, segundos = map(int, partes)
        return horas * 60 + minutos

    # Si el formato no es válido, retornar 0 o manejar el error como prefieras
    return 0
pacientes_b = {
    row['ID']: convertir_a_minutos(row['INICIO VENTANA'])
    for index, row in pacientes_sm_semana.iterrows()
}
pacientes_b
```
